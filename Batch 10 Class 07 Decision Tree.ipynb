```python
import pandas as pd
df = pd.read_csv ('smoking data.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df.region = le.fit_transform(df['region'])
df.sex = le.fit_transform(df['sex'])
```

```python
df.head()
```

```python
x = df.drop('smoker', axis=1)
```

```python
y = df['smoker']
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(x,y)
```

```python
clf.predict([[3, 19, 0, 27.900, 0]])   #region= southwest: 3, age=19, sex= female: 0, bmi= 27.900,children=0
##prediction should be 'yes'
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
x_features = x.columns
```

```python
tree.plot_tree(clf)
tree.plot_tree(clf, filled=True, rounded=True, feature_names=x_features)
```
